```python
'''
Libraries, etc.
'''
import os
import time
import numpy as np # linear algebra
import scipy as sp # more w/ numpy arrays
import pandas as pd # data processing
import tensorflow as tf # machine learning
import matplotlib.pyplot as plt # graph/plot
```

```python
'''
Constants, Hyperparameters, etc.
'''
INPUT_DIR = './fer2013/'
INPUT_FILE = 'fer2013.csv'
NUM_LABELS = 7
NUM_TRAIN = 28709
LABEL_DICT = {
    0:'Angry', 
    1:'Disgust', 
    2:'Fear', 
    3:'Happy', 
    4:'Sad', 
    5:'Surprise', 
    6:'Neutral'
}
BASE_SIZE = 48

#startTime = time.time()
```

```python
'''
Load & Parse Data into Training/Test Labels & Images
'''
data = np.array(pd.read_csv(INPUT_DIR + INPUT_FILE))
train_indx = data[:,2] == 'Training'
train_data = data[train_indx]
test_data = data[np.logical_not(train_indx)]

train_count = len(train_data)
train_labels = train_data[:,0]
train_pixels = train_data[:,1]
train_images = np.zeros((train_count, BASE_SIZE, BASE_SIZE))
for x, px in enumerate(train_pixels):
    train_images[x, :, :] = np.reshape(px.split(' '), (BASE_SIZE, BASE_SIZE))

test_count = len(test_data)
test_labels = test_data[:,0]
test_pixels = test_data[:,1]
test_images = np.zeros((test_count, BASE_SIZE, BASE_SIZE))
for x, px in enumerate(test_pixels):
    test_images[x, :, :] = np.reshape(px.split(' '), (BASE_SIZE, BASE_SIZE))

#print(train_images[1])
#print(train_images.shape)
#print(test_images[1])
#print(test_images.shape)
```

```python
'''
Visualize some images to check that they have been read in properly
'''
for x in range(1,26):
    plt.subplot(5, 5, x)
    plt.imshow(train_images[x])
    
plt.show()
```
